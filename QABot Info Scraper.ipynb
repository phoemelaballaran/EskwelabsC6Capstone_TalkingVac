# Question Answering Bot Information Scraper

## Import libraries

```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

## World Health Organization FAQs

```python
url = "https://www.who.int/emergencies/diseases/novel-coronavirus-2019/question-and-answers-hub"
src = requests.get(url, headers = {'User-Agent':'Mozilla/5.0 (X11; Linux)'})
soup = BeautifulSoup(src.content, 'html.parser')

questions = soup.find_all('a', class_ = 'sf-list-vertical__item')


#To append to the url 'https://www.who.int'
questions_link = []

for x in questions:
     questions_link.append(x['href'])
```

```python
questions_list = []
index = 0

while index < len(questions_link):
    url = f"https://www.who.int{questions_link[index]}"
    src = requests.get(url, headers = {'User-Agent':'Mozilla/5.0 (X11; Linux)'})
    soup = BeautifulSoup(src.content, 'html.parser')

    q = soup.find_all('a', class_ = 'sf-accordion__link')
    a = soup.find_all('div', class_ = 'sf-accordion__content')
    
    qindex = 0
    
    while qindex < len(q):
        q1 = q[qindex].text.strip()
        a1 = a[qindex].find('p').get_text().strip().replace('\xa0','')
        questions_list.append([q1,a1])
        
        qindex += 1
        
    print(f"Done with: {index}")    
    index += 1
```

```python
who_faq = pd.DataFrame(data = questions_list, columns =['Question','Answer'])
who_faq
```

```python
who_faq.to_csv('WHO FAQ.csv')
```

## Centers for Disease Control and Prevention FAQs

```python
url = "https://www.cdc.gov/coronavirus/2019-ncov/faq.html"
src = requests.get(url, headers = {'User-Agent':'Mozilla/5.0 (X11; Linux)'})
soup = BeautifulSoup(src.content, 'html.parser')

q = soup.find_all('div', class_ = 'card-header')
a = soup.find_all('div', class_ = 'card-body')
```

```python
cdc_faq = []
i = 4

while i < 106:
    q1 = q[i].find('span').get_text()
    a1 = a[i-1].get_text().replace('\xa0','').replace('\n','')
    
    cdc_faq.append([q1,a1])
    
    i+=1 
```

```python
cdc_faq = pd.DataFrame(data = cdc_faq, columns =['Question','Answer'])
```

```python
cdc_faq.to_csv('CDC FAQ.csv')
```
